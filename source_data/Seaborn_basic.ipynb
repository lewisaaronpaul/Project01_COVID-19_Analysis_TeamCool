```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.rcParams["figure.figsize"] = [10,8]
tips_data = sns.load_dataset("tips")
tips_data.head()
```

```python
tips_data.info()
```

## The Dist Plots
The dist plot, also known as the distributional plot, is used to
plot the histogram of data for a specific column in the dataset.
To plot a dist plot, you can use the distplot() function of the
Seaborn library. The name of the column for which you want
to plot a histogram is passed as a parameter to the distplot()
function. The following script plots the dist plot for the
total_bill column of the tips dataset.

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.distplot(tips_data["total_bill"])
```

```python
sns.distplot(tips_data['tip'], kde = False)
```

## Joint Plot
The joint plot is used to plot the histogram distribution of two
columns, one on the x-axis and the other on the y-axis. A
scatter plot is by default drawn for the points in the two
columns. To plot a joint plot, you need to call the jointplot()
function. The following script plots a joint plot for the
total_bill and tip columns of the tips dataset.

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data)
```

### The scatter plot can be replaced by a regression line in a joint plot.
To do so, you need to pass reg as the value for the kind
parameter of the jointplot() function.

```python
sns.jointplot(x='size', y='total_bill', data=tips_data, kind = 'reg')
```

## Pair Plot
The pair plot is used to plot a joint plot for all the
combinations of numeric and Boolean columns in a dataset.
To plot a pair plot, you need to call the pairplot() function and
pass it to your dataset.

```python
tips_data.head()
```

```python
sns.pairplot(data=tips_data)
```

## You can also plot multiple pair plots per value in a categorical column. 
To do so, you need to pass the name of the
categorical column as the value for the hue parameter. The
following script plots two pair plots (one for lunch and one for
dinner) for every combination of numeric or Boolean columns.

```python
sns.pairplot(data=tips_data, hue = 'time')
```

## 4.6. Bar Plot
The bar plot is used to capture the relationship between a
categorical and numerical column. For each unique value in a
categorical column, a bar is plotted, which by default, displays
the mean value for the data in a numeric column specified by
the bar plot.
In the following script, we first import the built-in Titanic
dataset from the Seaborn library via the load_dataset()
function.

```python
plt.rcParams["figure.figsize"] = [8,6]
sns.set_style('darkgrid')
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

## Next, we will call the barplot() function from the Seaborn
library to plot a bar plot that displays the average age of
passengers traveling in different classes of the Titanic ship.

```python
sns.barplot(x='pclass', y='age', data=titanic_data)
```

## You can further categorize the bar plot using the hue attribute.
For example, the following bar plot plots the
average ages of passengers traveling in different classes and
further categorized based on their genders.

```python
sns.barplot(x='pclass', y='age', hue ='sex', data=titanic_data)
```

## You can also plot multiple bar plots depending upon the
number of unique values in a categorical column. To do so,
you need to call the catplot() function and pass the
categorical column name as the value for the col attribute
column. The following script plots two bar plotsone for the
passengers who survived the Titanic accident and one for
those who didnt survive.

```python
sns.catplot(x='pclass', y='age', hue='sex', col='survived', data=titanic_data, kind='bar');
```

## Count Plot
The count plot plots plot like a bar plot. However, unlike the
bar plot, which plots average values, the count plot simply
displays the counts of the occurrences of records for each
unique value in a categorical column. The countplot() function
is used to plot a count plot with Seaborn. The following script
plots a count plot for the pclass column of the Titanic dataset.

```python
sns.countplot(x='pclass', data=titanic_data)
```

## Like a bar plot, you can also further categorize the count plot by passing a value for the hue parameter. 
The following script
plots a count plot for the passengers traveling in different
classes of the Titanic ship categorized further by their
genders.

```python
sns.countplot(x='pclass', hue='sex', data=titanic_data)
```

## Box Plot
The box plot is used to plot quartile information for data in a
numeric column. To plot a box plot, the boxplot() method is
used. To plot a horizontal box plot, the column name of the
dataset is passed to the x-axis. The following script plots a
box plot for the fare column of the Titanic dataset.

```python
sns.boxplot(x=titanic_data['fare'])
```

```python
#Similarly, the following script plots the vertical box plot for the fare column of the Titanic dataset.
sns.boxplot(y=titanic_data['fare'])
```

### You can also plot multiple box plots for every unique value in a categorical column. 
For instance, the following script plots
box plots for the age column of the passengers who traveled
alone as well as for passengers who were accompanied by at
least one other passenger.

```python
sns.boxplot(x='alone', y='age', data=titanic_data)
```

```python
sns.boxplot(x='alone', y='age', hue='sex', data=titanic_data)
```

```python
#By default, the box plot draws outliers. To remove outliersfrom a box plot, you have to pass False as the value for the
#showfliers attribute of the boxplot() function.
sns.boxplot(x='alone', y='age', hue='sex',data=titanic_data, showfliers=False)
```

## Violin Plot
Violin plots are similar to box plots. However, unlike box plots that plot quartile information, violin plots plot the overall
distribution of values in numeric columns. The following script plots two violin plots for the passengers traveling alone and
along with another passenger. The violinplot() function isused to plot a swarm plot with Seaborn.

The output shows that among the passengers traveling alone, the passengers whose age is less than 15 are very few as
shown by the orange violin plot on the right. This behavior is understandable as children are normally accompanied by
someone. This can be further verified by looking at the blue violin plot on the left that corresponds to the passengers
accompanied by other passengers.

The hue attribute can also be used to categorize the violinplot further. The following script plots a violin plot, first categorized by whether or not the passengers travel alone, and then by their gender.

```python
sns.violinplot(x='alone', y='age', data=titanic_data)
```

```python
sns.violinplot(x='alone', y='age', hue='sex',data=titanic_data)
```

```python
# For a better comparison and to save space, you can also plot split violin plots. In split violin plots, each half 
#corresponds to one value in a category column. For instance, the following script plots two violin plotsone each 
#for the passengers traveling alone and for the passengers not traveling alone.
#Each plot is further split into two parts based on the genders of the passengers.
sns.violinplot(x='alone', y='age', hue='sex',data=titanic_data, split=True)
```

## Scatter Plot
Lets now plot a scatter plot with the values from the total_bill column of the tips dataset on the x-axis and 
values from the tips column on the y-axis. To plot a scatter plot, you need to call the scatterplot() method 
of the Seaborn library.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```

```python
# To change the color of the scatter plot, simply pass the first letter of any color to the color attribute 
# of the scatterplot() function.
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'r')
```

```python
# Finally, to change the marker shape for the scatter plot, you need to pass a value for the marker attribute. 
# For example, the following scatter plot plots blue x markers on the scatter plot.
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## Styling Seaborn Plots
While Matplotlib plots might not be that aesthetically pleasing, you can create aesthetically pleasant plots with Seaborn. For
instance, you can set the background style of the plot with set_style() function. You can pass one of the darkgrid,whitegrid, white, and ticks as a value to the set_style() function. Here is an example of the set_style() function.

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

```python
# Similarly, if you pass whitegrid as a value to the set_style() function, you will see grids with a white 
# background. Here is an example.
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## In addition to styling the background, you can style the plot for different devices via the set_context() function. 
By default, the context is set to notebook. However, if you want to plot your plot on a poster, you can pass poster 
as a parameter to the set_context() function. In the output, you will see a plot with bigger annotations, as shown below.

```python
sns.set_context('poster')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## Heat Maps
A heat map is a type of matrix plot and is used to plot the correlation between numeric columns in a dataset in the form
of a matrix. To plot a heat map, both columns and rows should have meaningful information. To plot a heat map, we will use
the Titanic dataset. Lets first import the Titanic dataset.

```python
titanic_data.head()
```

```python
plt.rcParams['figure.figsize'] = [8,6]
sns.set_style('darkgrid')
```

```python
# The above output shows that we have meaningful information along columns only. To plot a heat map, we need meaningful
# information on rows as well. One of the ways to get meaningful information among rows is to call the corr() function on 
# a Pandas dataframe. Lets see what the corr() function does to the Pandas dataframe that contains the Titanic dataset.
titanic_data.corr()
```

#### From the output above, you can see that we now have meaningful information across rows as well. In the following script, we first increase the default plot size and then pass the correlation matrix of the Titanic dataset to the heatmap() function to create a heat map. You can see a heat map in the output, as shown below. The higher the correlation is, the darker the cell containing the correlation.

```python
plt.rcParams["figure.figsize"] = [14,9]
corr_values = titanic_data.corr()
sns.heatmap(corr_values, annot= True)
```

```python
# You can see that the above plot is cropped from the top and bottom. The following script plots the uncropped plot. 
# In the following script, we use the set_ylim() method to increase the plot size from top and bottom cell by 0.5 percent.
plt.rcParams['figure.figsize'] = [14,9]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
# You can also change the default color of the heatmap. To do so, you need to pass a value for the cmap attribute of the
# heatmap() function. Look at the script below:
sns.set_context()
plt.rcParams['figure.figsize'] = [14,9]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True, cmap = 'coolwarm')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```
