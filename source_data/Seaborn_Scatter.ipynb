```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Restaurant Data
plt.rcParams["figure.figsize"] = [10,8]
tips_data = sns.load_dataset("tips")
tips_data.head()
```

```python
# Titanic Data
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

## Scatter Plot
Letâ€™s now plot a scatter plot with the values from the total_bill column of the tips dataset on the x-axis and 
values from the tips column on the y-axis. To plot a scatter plot, you need to call the scatterplot() method 
of the Seaborn library.

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
```

```python
# To change the color of the scatter plot, simply pass the first letter of any color to the color attribute 
# of the scatterplot() function.
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'r')
```

```python
# Finally, to change the marker shape for the scatter plot, you need to pass a value for the marker attribute. 
# For example, the following scatter plot plots blue x markers on the scatter plot.
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## Styling Seaborn Plots
While Matplotlib plots might not be that aesthetically pleasing, you can create aesthetically pleasant plots with Seaborn. For
instance, you can set the background style of the plot with set_style() function. You can pass one of the darkgrid,whitegrid, white, and ticks as a value to the set_style() function. Here is an example of the set_style() function.

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

```python
# Similarly, if you pass whitegrid as a value to the set_style() function, you will see grids with a white 
# background. Here is an example.
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## In addition to styling the background, you can style the plot for different devices via the set_context() function. 
By default, the context is set to notebook. However, if you want to plot your plot on a poster, you can pass poster 
as a parameter to the set_context() function. In the output, you will see a plot with bigger annotations, as shown below.

```python
#sns.set_context('poster')
sns.scatterplot(x='total_bill', y='tip', data=tips_data, color = 'b', marker = 'x')
```

## Heat Maps
A heat map is a type of matrix plot and is used to plot the correlation between numeric columns in a dataset in the form
of a matrix. To plot a heat map, both columns and rows should have meaningful information. To plot a heat map, we will use
the Titanic dataset. Letâ€™s first import the Titanic dataset.

```python
titanic_data.head()
```

```python
plt.rcParams['figure.figsize'] = [8,6]
sns.set_style('darkgrid')
```

```python
# The above output shows that we have meaningful information along columns only. To plot a heat map, we need meaningful
# information on rows as well. One of the ways to get meaningful information among rows is to call the corr() function on 
# a Pandas dataframe. Letâ€™s see what the corr() function does to the Pandas dataframe that contains the Titanic dataset.
titanic_data.corr()
```

#### From the output above, you can see that we now have meaningful information across rows as well. In the following script, we first increase the default plot size and then pass the correlation matrix of the Titanic dataset to the heatmap() function to create a heat map. You can see a heat map in the output, as shown below. The higher the correlation is, the darker the cell containing the correlation.

```python
plt.rcParams["figure.figsize"] = [14,9]
corr_values = titanic_data.corr()
sns.heatmap(corr_values, annot= True)
```

```python
# You can see that the above plot is cropped from the top and bottom. The following script plots the uncropped plot. 
# In the following script, we use the set_ylim() method to increase the plot size from top and bottom cell by 0.5 percent.
plt.rcParams['figure.figsize'] = [14,9]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
# You can also change the default color of the heatmap. To do so, you need to pass a value for the cmap attribute of the
# heatmap() function. Look at the script below:
sns.set_context()
plt.rcParams['figure.figsize'] = [14,9]
corr_values = titanic_data.corr()
ax = sns.heatmap(corr_values, annot= True, cmap = 'coolwarm')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

In addition to changing the color of the heat map, you can also specify the line color and width that separates cells in a
heat map. Letâ€™s import the flights dataset from the Seaborn library. The flights dataset contains records of the passengers traveling each month from 1949 to 1960.

```python
flights_data = sns.load_dataset('flights')
flights_data
```

In addition to the corr() function, which returns the correlation between the numeric columns, you can also use the
pivot_table() function to get a matrix with meaningful rows and color. For instance, the pivot_table() function in the
following script returns a matrix where rows represent months, columns represent years from 1949 to 1960, and each
cell contains the number of passengers traveling in a specificmonth of a specific year. The matrix created via the
pivot_table() method can then be plotted via the heatmap() function.

```python
# Pivot Table
flights_data_pivot =flights_data.pivot_table(index='month', columns='year', values='passengers')
flights_data_pivot
```

```python
plt.rcParams['figure.figsize'] = [10,8]
ax = sns.heatmap(flights_data_pivot, cmap = 'coolwarm', linecolor='black', linewidth=1)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

## Pair Grids
In the previous chapter, you saw how the pair plot can be used to plot relationships between numeric columns of a dataset. Before we see pair grids in action, letâ€™s revise how the pair plot works. The following script plots the pair plot for the
tips dataset.

```python
tips_data.head()
```

```python
plt.rcParams['figure.figsize'] = [10,8]
sns.pairplot(tips_data)
```

Letâ€™s now plot a pair grid for the tips dataset. To do so, you have to pass the Pandas dataframe containing the tips dataset
to the PairGrid() function, as shown below. In the output, you should see the empty grids as follows.

```python
sns.PairGrid(tips_data)
```

To actually plot a graph on the grids returned by the PairGrid() function, you need to call the map() function on the object returned by the PairGrid() function. Inside the map function, the type of plot is passed as a parameter. For instance, the following PairGrid() function plots a scatter plot for all the pairs of numerical columns in the tips dataset.

```python
pairgrids = sns.PairGrid(tips_data)
pairgrids.map(plt.scatter)
```

With a pair grid, you can plot different types of plots on the diagonal, upper portion from the diagonal, and the lower
portion from a diagonal. For instance, the following pair grid plots a kernel density estimation plots on diagonal,
distributional plots on the upper part of the diagonal, and scatter plots on the lower side of the diagonal.

```python
pgrids = sns.PairGrid(tips_data)
pgrids.map_diag(sns.distplot)
pgrids.map_upper(sns.kdeplot)
pgrids.map_lower(plt.scatter)
```

## Regression Plots
Regression plots are used to plot a scatter plot for two columns along with the regression lines. To plot a regression
plot with Seaborn, you need to call the lmplot() method. The following script plots a regression plot for total_bill and tips
columns of the tips dataset.

```python
plt.rcParams['figure.figsize'] = [10,6]
sns.lmplot(x='total_bill', y='tip', data =tips_data)
```

You can plot regression plots for two columns on the y-axis. To do so, you need to pass a column name for the hue
parameter of the lmplot() function.

```python
sns.lmplot(x='total_bill', y='tip', data =tips_data, hue='time', markers=["o", "x"], palette="Set1")
```

In addition to having on plot for both the columns on the xaxis, you can create two different plots via col parameter as
shown below:

```python
sns.lmplot(x='total_bill', y='tip', data =tips_data, col='time')
```
