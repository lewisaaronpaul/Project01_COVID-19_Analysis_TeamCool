```python
import pandas as pd
import csv
import os
import us
import glob
```

```python
path = os.path.join('State_Files')
all_files = glob.glob(path + '/*.csv')

ind_files = []

for state in all_files:
    df = pd.read_csv(state, index_col=None, header=0)
    ind_files.append(df)

combined_df = pd.concat(ind_files, axis=0, ignore_index = True)
combined_df
```

```python
HRSA_Status = combined_df.loc[combined_df['data.HRSA.gov – HPSA Find'] == 'Primary Care']
HRSA_Status = HRSA_Status.rename(columns={"data.HRSA.gov – HPSA Find": "Discipline", "data.HRSA.gov – HPSA Find.1": "HPSA ID",
                   'data.HRSA.gov – HPSA Find.2': 'HPSA Name', 'data.HRSA.gov – HPSA Find.3': 'Designation Type',
                   'data.HRSA.gov – HPSA Find.4': 'State Name', 'data.HRSA.gov – HPSA Find.5': 'County Name',
                   'data.HRSA.gov – HPSA Find.6': 'HPSA FTE Short', 'data.HRSA.gov – HPSA Find.7': 'HPSA Score', 
                   'data.HRSA.gov – HPSA Find.8': 'Status', 'data.HRSA.gov – HPSA Find.9': 'Rural Status',
                   'data.HRSA.gov – HPSA Find.10': 'Designation', 'data.HRSA.gov – HPSA Find.11': 'Update Date'})
# HRSA_Status.to_csv('Test.csv', index=False)
HRSA_Status
```
